```python
from datetime import datetime
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
es_index = 'facebook'
es_post_doc_type = 'post'
es_comment_and_reaction_doc_type = 'action'
```

```python
# query all
page = es.search(
    index=es_index,
    scroll = '2m',
    size = 1000,
    body={"query": {"match_all": {}}})
print("Got %d Hits:" % page['hits']['total'])
sid = page['_scroll_id']
scroll_size = page['hits']['total']
  
# Start scrolling
while (scroll_size > 0):
    print("Scrolling...")
    page = es.scroll(scroll_id = sid, scroll = '2m')
    # Update the scroll ID
    sid = page['_scroll_id']
    # Get the number of results that we returned in the last scroll
    scroll_size = len(page['hits']['hits'])
    print("scroll size: " + str(scroll_size))
    # Do something with the obtained page
    for data in page['hits']['hits']:
        print(data)
        break
```

```python
# query intersect comments
page = es.search(
    index=es_index,
    doc_type=es_comment_and_reaction_doc_type,
    ignore=404, 
    scroll = '2m',
    size = 1000,
    body={"query": {
        "bool": {
            "must": [{
                "exists": {
                    "field": "message"
                }
            },
            {
                "exists": {
                    "field": "type"
                }
            }
            ]
        }
    }
})
print("Got %d Hits:" % page['hits']['total'])
sid = page['_scroll_id']
scroll_size = page['hits']['total']
  
# Start scrolling
while (scroll_size > 0):
    print("Scrolling...")
    page = es.scroll(scroll_id = sid, scroll = '2m')
    # Update the scroll ID
    sid = page['_scroll_id']
    # Get the number of results that we returned in the last scroll
    scroll_size = len(page['hits']['hits'])
    print("scroll size: " + str(scroll_size))
    # Do something with the obtained page
    for data in page['hits']['hits']:
        print(data)
        break
```
